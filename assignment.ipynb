```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
url = "https://www.musashino-u.ac.jp"
res  = requests.get(url) 
res.encoding = res.apparent_encoding

print(res.status_code)
print(res.text[:500])
```

```python
soup = BeautifulSoup(res.text, "html.parser")

# aタグを全部取得
links = soup.find_all("a")

print(f"リンクの数：{len(links)}件\n")

# 最初の10個だけ確認してみよう
for a in links[:10]:
    href = a.get("href")
    print(href)
```

```python
base = "https://www.musashino-u.ac.jp/"

internal_links = []

for a in links:
    href = a.get("href")
    if href and href.startswith(base):
        internal_links.append(href)
    elif href and href.startswith("/"):
        # 相対パス（/department/...）を絶対URLに変換
        internal_links.append(base.rstrip("/") + href)

print(f"学内リンクの数：{len(internal_links)}件\n")

for link in internal_links[:10]:
    print(link)
```

```python
for url in internal_links:
    try:
        time.sleep(1)
        r = requests.get(url)
        r.encoding = r.apparent_encoding  # ← これを追加
        if r.status_code == 200:
            soup = BeautifulSoup(r.text, "html.parser")
            title_tag = soup.find("title")
            title = title_tag.text.strip() if title_tag else "No Title"
            sitemap[url] = title
            print(f"✅ {url} -> {title}")
        else:
            print(f"⚠️ {url} -> ステータスコード: {r.status_code}")
    except Exception as e:
        print(f"❌ {url} -> エラー: {e}")

```
